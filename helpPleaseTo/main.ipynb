```python
import pandas as pandas
import numpy as numpy
import matplotlib.pyplot as pyplot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load data
data = pandas.read_json('dataSet.json')
data.head()
```

```python
x = data[['schoolNameLength', 'students_number', 'students_female_number', 'students_sexRatio', 'students_moveOutRatio', 'students_moveInRatio', 'generalTeacher_number', 'generalTeacher_studentsRatio', 'generalTeacher_female_number', 'generalTeacher_sexRatio', 'generalTeacher_male_studentsRatio', 'generalTeacher_female_studentsRatio', 'positionTeacher_number', 'positionTeacher_female_number', 'positionTeacher_sexRatio', 'schoolSize', 'schoolSize_studentsRatio', 'generalClass_number', 'specialClass_number', 'generalClass_studentsRatio', 'specialClass_studentsRatio']]

y = data[['employmentRate']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
# 모델
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
pyplot.scatter(y_test, y_predict)
pyplot.xlabel("Actual employmentRate")
pyplot.ylabel("Predicted employmentRate")
pyplot.title("MULTIPLE LINEAR REGRESSION")
pyplot.show()
```
